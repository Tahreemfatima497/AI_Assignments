```python
import matplotlib.pylab as plt
import seaborn as sb
```

```python
sb.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sb.barplot(x='Class', y='Fee', data=data, estimator=sum)

plt.show()
```

```python
# Check how many students pay same fee
import matplotlib.pylab as plt
import seaborn as sb
plt.figure(figsize=(10, 6))
sb.countplot(x="Fee", data=data, palette='tab10')

plt.show()
```

```python
# Visualization 2: Scatter plot of Roll.No vs Fee
plt.figure(figsize=(10, 6))
sb.scatterplot(x='Roll.No', y='Fee', data=data, hue='Class', palette='deep')
plt.title('Scatter Plot of Roll.No vs Fee')
plt.xlabel('Roll.No')
plt.ylabel('Fee')
plt.legend(title='Class')
plt.show()
```
